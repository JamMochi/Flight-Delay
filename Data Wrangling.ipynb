# Evaluate Data Type

```python
mod_flights.info()
```

Some of the columns datatype does not really represent what the data is trying to relay to us. In this section, we will convert the data in this dataframe into its appropriate datatype.

```python
categorical_data = ['DAY_OF_WEEK', 'FLIGHT_NUMBER']

for col in categorical_data:
    mod_flights[col] = mod_flights[col].astype('category')
```

```python
mod_flights['DATE'] = pd.to_datetime(mod_flights[['YEAR', 'MONTH', 'DAY']])
```

```python
mod_flights.iloc[3,:].T
```

```python
mod_flights[['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME']].head()
```

```python
def number_to_datetime(number):
    
    if pd.isnull(number):
        return np.nan
    
    number = int(number)
    if number == 2400:
        number = 0
    padded_number = f'{number:04}'
    return datetime.time(int(padded_number[0:2]), int(padded_number[2:4]))
```

```python
need_time_adjust = ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'SCHEDULED_ARRIVAL', 'ARRIVAL_TIME']

deepcopy = mod_flights.copy()

for adjust in need_time_adjust:
    deepcopy[adjust] = deepcopy[adjust].apply(number_to_datetime)
```
